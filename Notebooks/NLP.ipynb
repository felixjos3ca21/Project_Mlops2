```python
# listas de lo que parecen ser diccionarios, y listas vacías.
# procedemos a la transformación de texto a diccionario
import ast
import json

def convertir_a_dict(valor):
    datos_list = ast.literal_eval(valor)
    datos_str_list = [json.dumps(d) for d in datos_list]
    datos_dict_list = [json.loads(d) for d in datos_str_list]
    return datos_dict_list

# Aplicar la función a toda la columna 'cast' agregando una nueva columna al DF llamada 'cast_dict'
datos['reviews1'] = datos['reviews'].apply(convertir_a_dict)
```

```python
# Utiliza la función 'explode' para expandir la lista de diccionarios en la columna 'reviews1'
datos_expandidos = datos.explode('reviews1')

# Reinicia los índices del DataFrame
datos_expandidos.reset_index(drop=True, inplace=True)

# Define una función para extraer información de las claves de los diccionarios
def extraer_clave(diccionario, clave):
    try:
        return diccionario.get(clave, None) if isinstance(diccionario, dict) else None
    except KeyError:
        return None

# Extrae las claves 'posted', 'item_id', 'recommend' y 'review' en columnas separadas
datos_expandidos['posted'] = datos_expandidos['reviews1'].apply(extraer_clave, clave='posted')
datos_expandidos['item_id'] = datos_expandidos['reviews1'].apply(extraer_clave, clave='item_id')
datos_expandidos['recommend'] = datos_expandidos['reviews1'].apply(extraer_clave, clave='recommend')
datos_expandidos['review'] = datos_expandidos['reviews1'].apply(extraer_clave, clave='review')
```

```python
from langdetect import detect
import pandas as pd



# Define una función para detectar el idioma de una revisión
def detectar_idioma(review):
    try:
        return detect(review)
    except:
        return None

# Aplica la función a la columna 'review' para detectar el idioma y crea una nueva columna 'language'
datos_expandidos['language'] = datos_expandidos['review'].apply(detectar_idioma)

# Imprime el DataFrame resultante con la nueva columna 'language'
print(datos_expandidos['review'][0])
```

```python
import langid

def detect_language(text):
    try:
        return langid.classify(text)[0]
    except:
        return None

datos_expandidos['language1'] = datos_expandidos['review'].apply(detect_language)
```

```python
import re

def limpiar_texto(texto):
    # Convierte el texto a minúsculas
    texto = texto.lower()
    # Elimina espacios en blanco adicionales
    texto = ' '.join(texto.split())
    # Elimina saltos de línea
    texto = texto.replace('\n', ' ')
    # Elimina comas
    texto = texto.replace(',', '')
    # Elimina comillas
    texto = texto.replace('"', '')
    # Elimina signos de exclamación
    texto = texto.replace('!', '')
    # Elimina otros caracteres especiales si es necesario
    texto = re.sub(r'[^a-zA-Z0-9\s]', '', texto)
    return texto

# 4. Aplica la función a la columna
datos_expandidos['texto_review'] = datos_expandidos['review'].apply(limpiar_texto)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = nltk.corpus.stopwords.words('english')
```

```python
stopwords
```

```python
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer('\w+')
# Función de tokenización personalizada
def tokenize_if_english(row):
    if row['language1'] == 'en':
        return regexp.tokenize(row['texto_review'])
    else:
        return None

# Aplicar la función a la columna 'texto_review'
datos_expandidos['texto_token'] = datos_expandidos.apply(tokenize_if_english, axis=1)
```

```python
## Quitar Stopwords
# Función para quitar stopwords
def remove_stopwords(text):
    if text is None:
        return []  # Si el texto es None, retorna una lista vacía
    return [word for word in text if word not in stopwords]

# Aplicar la función a la columna 'texto_token'
datos_expandidos['texto_token_sinS'] = datos_expandidos['texto_token'].apply(remove_stopwords)
```

```python
datos_expandidos['texto_token_sinS']
```

```python
# Stemming

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

stemmer.stem('rainning')
```

```python
datos_expandidos['texto_token_stemming'] = datos_expandidos['texto_token_sinS'].apply(lambda x : [stemmer.stem(item) for item in x])
```

```python
datos_expandidos['texto_token_stemming']
```

```python
datos_expandidos['texto_token_listo'] = datos_expandidos['texto_token_stemming'].apply(lambda x : ' '.join([item for item in x])) 
```

```python
datos_expandidos['texto_token_listo']
```

```python
## Analisi de sentimiento
import nltk

# Descarga el recurso necesario
#nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

datos_expandidos['sentiment'] = datos_expandidos['texto_token_listo'].apply(lambda x : analyzer.polarity_scores(x)['compound']) 
```

```python
datos_expandidos['sentiment']
```

```python
import numpy as np

# Crear una nueva columna 'sentiment_analysis' basada en 'sentiment'
datos_expandidos['sentiment_analysis'] = np.where(datos_expandidos['sentiment'] == 0, 1, 
                                                  np.where(datos_expandidos['sentiment'] > 0, 2, 0))
```
