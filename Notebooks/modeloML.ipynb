```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('stopwords')

# Convertir todas las columnas de texto a minúsculas
data_ml['id'] = data_ml['id'].astype(str).str.lower()
data_ml['genres'] = data_ml['genres'].astype(str).str.lower()
data_ml['tags'] = data_ml['tags'].astype(str).str.lower()
data_ml['specs'] = data_ml['specs'].astype(str).str.lower()
data_ml['publisher'] = data_ml['publisher'].astype(str).str.lower()

# Eliminar caracteres no deseados en todas las columnas
data_ml['genres'] = data_ml['genres'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))
data_ml['tags'] = data_ml['tags'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))
data_ml['specs'] = data_ml['specs'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))
data_ml['publisher'] = data_ml['publisher'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))

# Eliminar palabras vacías (stop words) en todas las columnas
stop_words = set(stopwords.words('english'))
data_ml['genres'] = data_ml['genres'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))
data_ml['tags'] = data_ml['tags'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))
data_ml['specs'] = data_ml['specs'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))
data_ml['publisher'] = data_ml['publisher'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))
```

```python
data_ml['combined_text'] = data_ml['genres'] + ' ' + data_ml['tags'] + ' ' + data_ml['specs'] + ' ' + data_ml['publisher']
```

```python
data_ml = data_ml[['id','title','combined_text']]
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer(stop_words='english')
vectorized_data = vectorizer.fit_transform(data_ml['combined_text'])
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(vectorized_data)
```

```python
# Crear y ajustar el modelo KNN fuera de la función
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(vectorized_data)


def get_recomendations(id, modelo=knn_model):
    data_ml['id'] = data_ml['id'].astype(str)
    # Verificar si el id existe en el DataFrame
    if id not in data_ml['id'].values:
        return {'error': 'ID not found'}
    
    # Obtener índice del id
    index = data_ml[data_ml['id'] == id].index[0]
    
    # Numero de recomendaciones
    num_recomen = 5
    
    # Obtener recomendaciones basadas en el índice de consulta
    _, indices = knn_model.kneighbors(vectorized_data[index], n_neighbors=num_recomen+1)
    
    # Obtener índices de los id recomendadas
    index_id = indices.flatten()[1:]
    
    # Devolver una estructura de diccionario con los IDs de los title recomendadas
    result = {'lista recomendada': data_ml['title'].iloc[index_id].tolist()}

    return result
```

```python
id = input('Ingrese un id: ')
recomendacion = get_recomendations(id)
print(recomendacion)
```
